```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "data.csv"
df = pd.read_csv(path)
df
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
df.drop(columns = ['id','Unnamed: 32'],inplace=True)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,1:]
Y = df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size = 0.66)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train.iloc[:,:] = scaler.fit_transform(X_train.iloc[:,:])
X_test.iloc[:,:] = scaler.transform(X_test.iloc[:,:])
```

```python
X_train['diagnosis'] = Y_train
X_test['diagnosis'] = Y_test
X_train.to_csv('BC_train.csv',header=True,index=False)
X_test.to_csv('BC_test.csv',header=True,index=False)
```

```python
corr = X_train.corr()
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
mask = np.triu(np.ones_like(corr,dtype=bool))

f,ax = plt.subplots(figsize=(12,10))

sns.heatmap(corr,mask=mask,cmap= 'RdBu')
plt.show()
```
